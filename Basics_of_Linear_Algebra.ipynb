Matrix Power

```python
x=np.random.random([3,3])
 # positive powers of matrix
x_2=np.linalg.matrix_power(x,2)
x_4=np.linalg.matrix_power(x,4)
 # inverse of matrix
x_inv_1 = np.linalg.matrix_power(x,-1)
x_inv_3 = np.linalg.matrix_power(x,-3)
print('matrix \n',x)
print('\n matrix to the power 2\n',x_2)
print('\n matrix to the power 4\n',x_4)
print('\n matrix inverse \n',x_inv_1)
print('\n matrix cubic inverse \n',x_inv_3)
```

Eigen Values And Eigen Vectors

```python
x = np.arange(9).reshape(3,3)
eig_val, eig_vec = np.linalg.eig(x)
print('Eigenvalues are: \n', eig_val)
print('\nEigenvectors are: \n', eig_vec)
```

```python
# Eigenvalues alone can be determined using the method eigvals as shown below.

x = np.arange(9).reshape(3,3)
eigenvalues = np.linalg.eigvals(x)
print(eigenvalues)
print(x)
```

System of Equations

```python
# Coefficient Matrix
a = np.random.randint(1,30,[4,4])
# Dependent variable vector
b = np.array([1,3,5,7])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
# Check for correctness
B = a.dot(x)
print(B)
```

Singular Value Decomposition

```python
from skimage import data
# download a sample image
image = data.astronaut()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# Normalize the image by dividing it by the maximum value, 255 and reorder the axes to be (3,400,600).

# normalize image
img = image/255.0
# reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2,0,1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
# S matrix should have dimensions suitable for matrix multiplication
Sigma = np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
#size 640 is different from 427 was the error if we use rocket
```

```python
# NumPy SVD reconstruction

# Reconstruct the data by reducing the common dimensions from 512 to 50.

k = 50
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

Dimensionality Reduction

```python
k = 30
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
